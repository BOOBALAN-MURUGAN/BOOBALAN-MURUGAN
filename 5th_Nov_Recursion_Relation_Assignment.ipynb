1. Find the value of T(2) for the recurrence relation T(n) = 3T(n-1) + 12n, given that T(0)=5.

    Recurrence relation:
    T(n) = 3T(n-1) + 12n
    
    Base Condition:
    T(0)=5
    
    Calculating the T(2) will be:
    
    Here T(0) value is 5 has been defined hence using for substitution
    T(1) = 3T(0)+12*1 --> 3*5 +12 = 27 #Then T(1) = 27
    
    
    T(2) = 3T(1)+(12*2) --> 3*27 + 24 = 105
    
    Finally T(2) value would be 105
